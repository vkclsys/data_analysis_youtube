```python
import numpy as np
import pandas as pd 
import matplotlib.pylab as plt
import seaborn as sns
plt.style.use('ggplot')
import plotly.express as px
```

```python
youtube_data=pd.read_csv("Global YouTube Statistics.csv", encoding="latin-1",index_col = "rank")
```

```python
# Drop rows with missing values
youtube_data.dropna(inplace=True)
# Remove duplicates
youtube_data.drop_duplicates(inplace=True)

# Convert data types if needed
youtube_data['subscribers'] = youtube_data['subscribers'].astype(int)
```

```python
from matplotlib.ticker import FuncFormatter
def millions(x, pos):
    'The two args are the value and tick position'
    return '%1.1fM' % (x * 1e-6)


formatter = FuncFormatter(millions)
```

```python
import matplotlib.pyplot as plt

# Group by category and sum the subscribers
top_10_channels = youtube_data.groupby('category')['subscribers'].sum().nlargest(10)

# Plot the top 10 channels

ax=top_10_channels.plot(kind='bar',x='category',y='Subscribers',title="Top 10 Youtube Channels in category",)
ax.yaxis.set_major_formatter(formatter)
```
